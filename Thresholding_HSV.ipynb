```python
import cv2
import numpy as np
import skimage

# Libreria para graficar
from matplotlib import pyplot as plt
```

```python
image = cv2.imread('imgs/leaf.jpg') 

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(20,10)) 
plt.imshow(image)           
plt.title('Leaf')
plt.axis('off')             
plt.show()                  
```

```python
image = cv2.imread('imgs/leaf.jpg') 

# Pasamos la imagen a espacio HSV
image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Para umbralizar o segmentar se debe escojer un nivel minimo y un maximo
min_val = (0,0,0)
max_val = (180,255,255) # El valor maximo del canal H es 180 grados

image = cv2.inRange(image, min_val, max_val)

plt.figure(figsize=(20,10)) 
plt.imshow(image, cmap='gray')           
plt.title('Segmented')
plt.axis('off')             
plt.show()      
```

```python
image = cv2.imread('imgs/leaf.jpg') 

# Pasamos la imagen a espacio HSV
image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Para umbralizar o segmentar se debe escojer un nivel minimo y un maximo
min_val = (35,0,0)
max_val = (60,255,255) 

# Para segmentar utilizamos la funcion de OpenCV inRange
image = cv2.inRange(image, min_val, max_val)

plt.figure(figsize=(20,10)) 
plt.imshow(image, cmap='gray')           
plt.title('Segmented')
plt.axis('off')             
plt.show()      

# Para guardar la image utilizamos el comando imwrite de OpenCV
cv2.imwrite('imgs/leaf_segmented.jpg', image)
```
