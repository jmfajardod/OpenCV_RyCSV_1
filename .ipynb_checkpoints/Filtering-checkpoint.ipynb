```python
import cv2
import numpy as np
import skimage

# Libreria para graficar
from matplotlib import pyplot as plt
```

```python
image = cv2.imread('imgs/leaf.jpg') 

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(20,10)) 
plt.imshow(image)           
plt.title('Leaf')
plt.axis('off')             
plt.show()                  
```

```python
image = cv2.imread('imgs/leaf.jpg') 

# Pasamos la imagen a espacio HSV
image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Para umbralizar o segmentar se debe escojer un nivel minimo y un maximo
min_val = (35,0,0)
max_val = (60,255,255) 

# Para segmentar utilizamos la funcion de OpenCV inRange
image = cv2.inRange(image, min_val, max_val)

plt.figure(figsize=(20,10)) 
plt.imshow(image, cmap='gray')           
plt.title('Segmented')
plt.axis('off')             
plt.show()      

# Para guardar la image utilizamos el comando imwrite de OpenCV
cv2.imwrite('imgs/leaf_segmented.jpg', image)
```

```python
image = cv2.imread('imgs/leaf.jpg') 

# Pasamos la imagen a espacio HSV
image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Segmentacion
min_val = (35,0,0)
max_val = (60,255,255) 
image = cv2.inRange(image, min_val, max_val)

# Filtro no lineal de mediana
image = cv2.medianBlur(image,3) # El tamaño del kernel debe ser impar

plt.figure(figsize=(20,10)) 
plt.subplot(2, 3, 1)
plt.imshow(image, cmap='gray')           
plt.title('Median filter, kernel=3x3')
plt.axis('off') 

image = cv2.medianBlur(image,5)

plt.subplot(2, 3, 2)
plt.imshow(image, cmap='gray')           
plt.title('Median filter, kernel=5x5')
plt.axis('off') 

image = cv2.medianBlur(image,9)

plt.subplot(2, 3, 3)
plt.imshow(image, cmap='gray')           
plt.title('Median filter, kernel=9x9')
plt.axis('off') 

image = cv2.medianBlur(image,15)

plt.subplot(2, 3, 4)
plt.imshow(image, cmap='gray')           
plt.title('Median filter, kernel=15x15')
plt.axis('off') 

image = cv2.medianBlur(image,25)

plt.subplot(2, 3, 5)
plt.imshow(image, cmap='gray')           
plt.title('Median filter, kernel=25x25')
plt.axis('off') 

image = cv2.medianBlur(image,51)

plt.subplot(2, 3, 6)
plt.imshow(image, cmap='gray')           
plt.title('Median filter, kernel=51x51')
plt.axis('off') 

plt.show()
```

```python
from skimage.morphology import erosion, dilation, opening, closing
from skimage.morphology import disk

image = cv2.imread('imgs/leaf.jpg') 

# Pasamos la imagen a espacio HSV
image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Segmentacion
min_val = (35,0,0)
max_val = (60,255,255) 
image = cv2.inRange(image, min_val, max_val)

# Para la operacion morfologica es necesario utilizar un elemento estructurante 
# en este caso vamos a utilizar un disco 
kernel = disk(1)
print("Kernel: \n", kernel)

# Se decidio utilizar un filtro morfologico de cierre para eliminar el ruido
filtered = closing(image, kernel)

plt.figure(figsize=(20,10)) 

plt.subplot(2, 3, 1)
plt.imshow(filtered, cmap='gray')           
plt.title('Filtered with disk of radius 1 px')
plt.axis('off')     

kernel = disk(3)
filtered = closing(image, kernel)
plt.subplot(2, 3, 2)
plt.imshow(filtered, cmap='gray')           
plt.title('Filtered with disk of radius 3 px')
plt.axis('off') 

kernel = disk(7)
filtered = closing(image, kernel)
plt.subplot(2, 3, 3)
plt.imshow(filtered, cmap='gray')           
plt.title('Filtered with disk of radius 7 px')
plt.axis('off') 

kernel = disk(15)
filtered = closing(image, kernel)
plt.subplot(2, 3, 4)
plt.imshow(filtered, cmap='gray')           
plt.title('Filtered with disk of radius 15 px')
plt.axis('off') 

kernel = disk(20)
filtered = closing(image, kernel)
plt.subplot(2, 3, 5)
plt.imshow(filtered, cmap='gray')           
plt.title('Filtered with disk of radius 20 px')
plt.axis('off') 

kernel = disk(30)
filtered = closing(image, kernel)
plt.subplot(2, 3, 5)
plt.imshow(filtered, cmap='gray')           
plt.title('Filtered with disk of radius 30 px')
plt.axis('off') 

kernel = disk(50)
filtered = closing(image, kernel)
plt.subplot(2, 3, 6)
plt.imshow(filtered, cmap='gray')           
plt.title('Filtered with disk of radius 50 px')
plt.axis('off') 

plt.show() 
```

```python
image = cv2.imread('imgs/leaf.jpg') 

# Pasamos la imagen a espacio HSV
image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Segmentacion
min_val = (35,0,0)
max_val = (60,255,255) 
image = cv2.inRange(image, min_val, max_val)

# Se pueden combinar diferentes tipos de filtro
filtered = cv2.medianBlur(image,9)

plt.figure(figsize=(20,10)) 
plt.subplot(1, 2, 1)
plt.imshow(filtered, cmap='gray')           
plt.title('Filtered with Median Filter')
plt.axis('off')     

kernel = disk(3)
filtered = opening(filtered, kernel)
plt.subplot(1, 2, 2)
plt.imshow(filtered, cmap='gray')           
plt.title('Filtered ')
plt.axis('off')   

plt.show() 
```
